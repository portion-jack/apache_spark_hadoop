```python
from pyspark import SparkContext
```

```python
sc = SparkContext('local')
```

```python
# list to rdd : parellelize
dt = [(1,2),(3,4),(5,6),(1,7),(3,4)]
nrdd = sc.parallelize(dt)
nrdd.collect()
```

```python
# reduce => action_func
# reduceByKey => transformation func
newrdd=nrdd.reduceByKey(lambda a,b : a+b)
newrdd.collect()
```

```python
newrdd1=nrdd.flatMap(lambda v : v)
newrdd1.collect()
```

```python
newrdd2 = nrdd.flatMap(lambda v:v).map(lambda v:(v,1))
newrdd2.collect()
```

```python
# action func
newrdd2.countByKey()
```

```python
newrdd2.countByValue()
```

```python
# transformation fuc
newrdd2.sortByKey().collect() 
```

```python
# transformation fuc
newrdd2.sortByKey(ascending=False).collect() 
```

```python
nrdd.sortBy(lambda v:v[1]).collect()
```
